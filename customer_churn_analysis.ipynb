```python
def converter(value):
    if value == 1:
        return "Yes"
    else:
        return "No"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(converter)
```

```python
ax = sns.countplot(x = df["Churn"] , data = df)

ax.bar_label(ax.containers[0])
plt.title("Count of customer by churn")
plt.show
```

```python
plt.figure(figsize = (3,3))
gb = df.groupby("Churn").agg({"Churn" : "count"})
plt.pie(gb["Churn"] , labels = gb.index , autopct = "%1.2f%%")
plt.title("Churn Percentage")
plt.show()
```

#we can see that 26.54% of out customers has been churned out so lets walk through the reason behind it.

```python
plt.figure(figsize = (7,3))
cg = sns.countplot(x = "gender" , data = df , hue = "Churn")
cg.bar_label(cg.containers[1])
plt.title("Churn By Gender")
plt.show()
```

```python
gender_churn = pd.crosstab(df['gender'], df['Churn'])
gender_churn_percent = gender_churn.div(gender_churn.sum(axis=1), axis=0) * 100

# Step 2: Plot stacked bar chart
fig, ax = plt.subplots(figsize=(7,3))
colors = ['#1f77b4', '#ff7f0e']  # You can customize this

bottom = np.zeros(len(gender_churn_percent))
for i, col in enumerate(gender_churn_percent.columns):
    values = gender_churn_percent[col].values
    bars = ax.bar(gender_churn_percent.index, values, bottom=bottom, color=colors[i], label=col)
    
    # Add labels
    for bar, percent in zip(bars, values):
        if percent > 0:
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_y() + bar.get_height()/2,
                    f'{percent:.1f}%', ha='center', va='center', fontsize=9)
    bottom += values

# Final touches
ax.set_title("Churn by Gender (as %)")
ax.set_ylabel("Percentage")
ax.legend(title="Churn")
plt.tight_layout()
plt.show()
```

```python
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Step 2: Calculate percentages
ct_percent = ct.div(ct.sum(axis=1), axis=0) * 100

# Step 3: Plot stacked bar chart
ax = ct_percent.plot(kind='bar', stacked=True, figsize=(7,3), color=['#1f77b4', '#ff7f0e'])  # example custom colors


# Step 4: Add percentage labels
for i, row in enumerate(ct_percent.values):
    bottom = 0
    for j, val in enumerate(row):
        if val > 0:
            ax.text(i, bottom + val/2, f'{val:.1f}%', ha='center', va='center', fontsize=9)
            bottom += val

# Step 5: Final touches

plt.title('Churn by SeniorCitizen (as %)')
plt.ylabel('Percentage')
plt.xlabel('SeniorCitizen')
plt.legend(title='Churn', bbox_to_anchor = (0.9 , 0.9))
                                           
plt.tight_layout()
plt.show()
```

```python
categorical_cols = [
    'gender', 'SeniorCitizen', 'Partner', 'Dependents',
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
    'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
    'PaymentMethod'
]

n_cols = 3
n_rows = (len(categorical_cols) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 4))
axes = axes.flatten()

for i, col in enumerate(categorical_cols):
    ax = axes[i]
    
    # Create cross-tab and percentage
    data = pd.crosstab(df[col], df['Churn'])
    data_percent = data.div(data.sum(axis=1), axis=0) * 100

    # Plot
    bottom = np.zeros(len(data_percent))
    categories = data_percent.columns
    x = np.arange(len(data_percent))

    for j, cat in enumerate(categories):
        values = data_percent[cat].values
        bars = ax.bar(data_percent.index, values, bottom=bottom, label=cat)
        
        # Add labels
        for bar, val in zip(bars, values):
            if val > 0:
                ax.text(bar.get_x() + bar.get_width()/2, bar.get_y() + bar.get_height()/2,
                        f'{val:.1f}%', ha='center', va='center', fontsize=8)
        bottom += values

    ax.set_title(f'Churn by {col}')
    ax.set_ylabel('Percentage')
    ax.tick_params(axis='x', rotation=30)

# Hide extra axes
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Shared legend
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, title='Churn', loc='upper center', ncol=2, fontsize=10)

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
